```python
import pandas as pd

data = pd.read_csv("data.csv")

data

import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(data['Date'], data['Temperature'])

plt.show()
```

```python
fig, ax = plt.subplots()

ax.pie([data['dept_1 sales'].sum(), data['dept_2 sales'].sum(), data['dept_3 sales'].sum()], labels=['dept_1 sales', 'dept_2 sales','dept_3 sales'])

ax.set_title('Sales by department')

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(data['Day type'], data['dept_3 sales'])
ax.plot(data['Day type'], data['dept_2 sales'])
ax.plot(data['Day type'], data['dept_1 sales'])
plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(data['Day type'],data['dept_1 sales'])


ax.set_title('Sales by Day type for Department 1')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(data['Day type'],data['dept_2 sales'])
ax.set_title('Sales by Day type for Department 2')

```

```python
fig, ax = plt.subplots()
ax.bar(data['Day type'],data['dept_3 sales'])
ax.set_title('Sales by Day type for Department 3')
```

```python
plt.bar(x=data['Day type'], height = data['MarkDown1'], label='MD1')
plt.bar(x=data['Day type'], height = data['MarkDown2'], label='MD2')
plt.bar(x=data['Day type'], height = data['MarkDown3'], label='MD3')
plt.bar(x=data['Day type'], height = data['MarkDown4'], label='MD4')
plt.bar(x=data['Day type'], height = data['MarkDown5'], label='MD5')
plt.title('Markdowns based on Day type')
plt.xlabel('Day type')
plt.ylabel('Markdown')
plt.legend(loc='upper left')
```
